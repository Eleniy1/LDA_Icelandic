```python
import seaborn as sns
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
#Соотношение тем для мужчин
sns.set_theme(style="ticks")

f, ax = plt.subplots(figsize=(10, 7))
sns.despine(f)

sns.histplot(
    df9[df9.sex=='karlmaður'],
    x="Dominant_Topic", hue="sex",
    multiple="stack",
    palette="light:m_r",
    edgecolor=".3",
    linewidth=.5,
)
plt.savefig('msex_theme.png')
```

```python
#соотношение тем для женщин
sns.set_theme(style="ticks")

f, ax = plt.subplots(figsize=(10, 7))
sns.despine(f)

sns.histplot(
    df9[df9.sex=='kvenmaður'],
    x="Dominant_Topic", hue="sex",
    multiple="stack",
    palette="light:m_r",
    edgecolor=".3",
    linewidth=.5,
)
plt.savefig('fsex_theme.png')
```

```python
#распределение топиков согласно количеству текстов и полу автора
sns.set(rc={'figure.figsize':(20, 10)})
sns.histplot(data=df9, x="sex", binwidth=0.05, hue="Dominant_Topic")
plt.savefig('count_sex_theme.png')
```

```python
#распределение тем для исландских текстов
sns.set(rc={'figure.figsize':(10, 5)})
sns.histplot(df8["Dominant_Topic"], binwidth=0.05)
plt.savefig('themes_is.png')
```

```python
import numpy as np
```

```python
#график заивимостей между полом автора, возрастом и выбором темы
pivot = df9.pivot_table(
    index=["Dominant_Topic"],
    columns=["sex"],
    values="age",
    aggfunc=np.average)
sns.set(rc={'figure.figsize':(20, 10)})
sns.heatmap(pivot)
plt.savefig('age_sex_theme.png')
```
